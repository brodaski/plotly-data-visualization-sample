# PSE:SM Candlestick Chart

Dataset: [PSE Dataset](https://www.kaggle.com/datasets/shanemaglangit/philippines-stock-exchange-dataset)

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df = pd.read_csv('stocks.csv')
df.head()
```

```python
df.drop(df[df['Code'] != 'SM'].index, inplace=True)
df.head()
```

```python
df.info()
```

```python
def price_to_float(string):
  string = string.replace(',', '')
  return float(string)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.drop(columns=['Volume', 'Change%'])
for col in ['Price', 'Open', 'High', 'Low']:
  df[col] = df[col].apply(price_to_float)
df.info()
```

```python
df.sort_values(by='Date', inplace=True)
df.head()
```

```python
df['Close'] = df['Open'].shift(-1)
df['Close'].iloc[-1] = df['Open'].iloc[-1]
df.head()
```

```python
df.tail()
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
fig.show()
```

```python
fig.write_html("sm-2011-2021-candlestick.html")
```

# WHR Choropleth Map

Data Source: <a href="https://www.kaggle.com/datasets/unsdsn/world-happiness?select=2019.csv">           Kaggle</a>

```python
!pip install pycountry
```

```python
import pandas as pd
import pycountry
import plotly.graph_objects as go
```

```python
df = pd.read_csv('2019whr.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=df.columns[3:9])
df = df.rename(columns={'Country or region': 'Country'})
df.info()
```

```python
def get_iso3(country_name):
  try:
    return pycountry.countries.lookup(country_name).alpha_3
  except LookupError:
    return None

df['Code'] = df['Country'].apply(get_iso3)
df.head()
```

```python
df.info()
```

```python
df.loc[df['Code'].isna() == True]
```

In order: TTO, XKX, None, RUS, TUR, CIV, COG, PSE, COD, SWZ

```python
codes_to_fill = ['TTO', 'XKX', None, 'RUS', 'TUR', 'CIV', 'COG', 'PSE', 'COD', 'SWZ']
na_indices = df.index[df['Code'].isna()]
for idx, code in zip(na_indices, codes_to_fill):
  df.at[idx, 'Code'] = code
df.info()
```

```python
fig = go.Figure(data=go.Choropleth(
    locations = df['Code'],
    z = df['Score'],
    text = df['Country'],
    colorscale = 'YlOrRd',
    autocolorscale=False,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title = 'Happiness Score',
))

fig.update_layout(
    title_text='2019 World Happiness Report',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Data Source: <a href="https://www.kaggle.com/datasets/unsdsn/world-happiness?select=2019.csv">\
            Kaggle</a>',
        showarrow = False
    )]
)

fig.show()
```

```python
fig.write_html("2019whr.html")
```

# SEA World Bank Data GDP Per Capita Bar Race Animation

Datasource: [World Bank](https://data.worldbank.org/indicator/NY.GDP.PCAP.CD)

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('gdp.pcap.csv')
df.head()
```

```python
sea_countries = ["BRN", "KHM", "IDN", "LAO", "MYS", "MMR", "PHL", "SGP", "THA", "TLS", "VNM"]
df = df.loc[df['Country Code'].isin(sea_countries)]
df.drop(columns=['Indicator Code', '2025'], inplace=True)
df.rename(columns={'Country Name': 'Country', 'Country Code': 'Code', 'Indicator Name': 'Text'}, inplace=True)
df.head()
```

```python
df = df.melt(
    id_vars=["Country", "Code", "Text"],
    var_name="Year",
    value_name="GDP.PCAP"
)

df.info()
```

```python
df.head()
```

```python
df.describe()
```

Bar Race Animation wherein there is a set range

```python
fig = px.bar(
    df,
    x='GDP.PCAP',
    y='Country',
    color='Country',
    orientation='h',
    animation_frame='Year',
    animation_group='Country',
    range_x=[0, 100000]
)

fig.show()
```

```python
fig.write_html("wb_gdppcap.html")
```

Bar Race Animation autorange

```python
fig = px.bar(
    df,
    x='GDP.PCAP',
    y='Country',
    color='Country',
    orientation='h',
    animation_frame='Year',
    animation_group='Country'
)

fig.update_layout(xaxis_autorange=True)

fig.show()
```

```python
fig.write_html("wb_gdppcap-autorange.html")
```
